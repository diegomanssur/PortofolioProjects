# Sleep, Health and Lifestyle

## <b>Introduction</b>
The Sleep Health and Lifestyle Dataset comprises 374 rows and 13 columns, covering a wide range of variables related to sleep and daily habits. It includes details such as gender, age, occupation, sleep duration, quality of sleep, physical activity level, stress levels, BMI category, blood pressure, heart rate, daily steps, and the presence or absence of sleep disorders. 

It's important to clarify that the data used is synthetic and created for illustrative purposes.

### <br>Key Features of the Dataset:
<b>Comprehensive Sleep Metrics:</b> Explore sleep duration, quality, and factors influencing sleep patterns.
<br><b>Lifestyle Factors:</b> Analyze physical activity levels, stress levels, and BMI categories.
<br><b>Cardiovascular Health:</b> Examine blood pressure and heart rate measurements.
<br><b>Sleep Disorder Analysis:</b> Identify the occurrence of sleep disorders such as Insomnia and Sleep Apnea.

<b>Dataset Columns:</b>
<br><b>Person ID:</b> An identifier for each individual.
<br><b>Gender:</b> The gender of the person (Male/Female).
<br><b>Age:</b> The age of the person in years.
<br><b>Occupation:</b> The occupation or profession of the person.
<br><b>Sleep Duration</b> (hours): The number of hours the person sleeps per day.
<br><b>Quality of Sleep</b> (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10.
<br><b>Physical Activity Level (minutes/day):</b> The number of minutes the person engages in physical activity daily.
<br><b>Stress Level (scale: 1-10):</b> A subjective rating of the stress level experienced by the person, ranging from 1 to 10.
<br><b>BMI Category:</b> The BMI category of the person (e.g., Underweight, Normal, Overweight).
<br><b>Blood Pressure (systolic/diastolic):</b> The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.
<br><b>Heart Rate (bpm):</b> The resting heart rate of the person in beats per minute.
<br><b>Daily Steps:</b> The number of steps the person takes per day.
<br><b>Sleep Disorder:</b> The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

## <b>Occupation Breakdown</b>

Our table contains data from 374 distinct individuals, divided in 11 different occupations. In order to understand and find insights related to these professions, we'll calculate the average for each factor.

### Average Sleep Duration per Occupation

SELECT Occupation, 
<br>ROUND(AVG(`Sleep Duration`),2) AS Average_Sleep_per_Occupation
<br>FROM `sleep_lifestyle_health`
<br>GROUP BY Occupation
<br>ORDER BY Average_Sleep_per_Occupation DESC
<br>;

### Average Quality of Sleep per Ocupation

SELECT Occupation, 
<br>ROUND(AVG(`Quality of Sleep`),2) AS Average_Quality_Sleep
<br>FROM `sleep_lifestyle_health`
<br>GROUP BY Occupation
<br>ORDER BY Average_Quality_Sleep DESC
<br>;

### Average Daily Steps per Occupation

SELECT Occupation, 
<br>ROUND(AVG(`Daily Steps`),2) AS Average_Daily_Steps
<br>FROM `sleep_lifestyle_health`
<br>GROUP BY Occupation
<br>ORDER BY Average_Daily_Steps DESC
<br>;

### Average Stress Level per Occupation

SELECT Occupation, 
<br>ROUND(AVG(`Stress Level`),2) AS Average_Stress_Level
<br>FROM `sleep_lifestyle_health`

<br>ORDER BY Average_Stress_Level DESC
<br>;

## <b>Sleep Disorders per BMI Cateegory</b>

The Body Mass Index (BMI) is a person's weight in kilograms divided by the square of height in meters. A high BMI can indicate high body fatness. BMI screens for weight categories that may lead to health problems, but it does not diagnose the body fatness or health of an individual. By grouping the BMI with the sleep disorder column, we can see if there's any relation between the two and understand the data better.

<b>Details about Sleep Disorder Column:</b>
<br><b>None:</b> The individual does not exhibit any specific sleep disorder.
<br><b>Insomnia:</b> The individual experiences difficulty falling asleep or staying asleep, leading to inadequate or poor-quality sleep.
<br><b>Sleep Apnea:</b> The individual suffers from pauses in breathing during sleep, resulting in disrupted sleep patterns and potential health risks.

### BMI Category and Insomnia

SELECT `BMI Category`, COUNT(`Sleep Disorder`) AS Insomnia
<br>FROM `sleep_lifestyle_health`
<br>WHERE `Sleep Disorder` = 'Insomnia'
<br>GROUP BY `BMI Category`
<br>;

### BMI Category and Sleep Apnea

SELECT `BMI Category`, COUNT(`Sleep Disorder`) AS Sleep_Apnea
<br>FROM `sleep_lifestyle_health`
<br>WHERE `Sleep Disorder` = 'Sleep Apnea'
<br>GROUP BY `BMI Category`
<br>;

### BMI Category and No Sleep Disorder

SELECT `BMI Category`, COUNT(`Sleep Disorder`) AS None
<br>FROM `sleep_lifestyle_health`
<br>WHERE `Sleep Disorder` = 'None'
<br>GROUP BY `BMI Category`
<br>;

## <b>Findings</b>
### <b>Occupation Breakdown</b>

As seen in the visualizations, <b>Engineers</b>, <b>Accountants</b>, <b>Lawyers</b> and <b>Nurses</b> have the best <b>Sleep Quality</b>, with an average between <b>7.9</b> and <b>8.5</b>, and an average between <b>7.1</b> and <b>8</b> hours on <b>Sleep Duration</b>. 

On the other hand, <b>Sales Representatives</b>,<b>Scientists</b> and <b> Sales People</b> are the professions that suffers from sleeping issues the most, with an average between <b>4</b> and <b>6</b> on <b>Sleep Quality</b>, and an average between <b>5.9</b> and <b>6.4</b> hours on <b>Sleep Duration</b>.

As for <b>Daily Steps</b>, <b>Lawyers</b>,<b>Nurses</b> and <b>Accountants</b> are the top 3 most active of the group with an average daily steps between <b>6800</b> and <b>8000</b>.
<br>The bottom 3 in this case are the <b>Sales Representatives</b>, <b>Scientists</b> and <b>Managers</b>, with an average daily steps between <b>3000</b> and <b>5500</b>.

At last, the top 3 most stressful occupations are <b>Sales Representatives</b>, <b>Scientists</b> and <b> Sales People</b>, with an average <b>Stress Level</b> between <b>6.7</b> and <b>8</b>.

The 3 least stressfull occupations are <b>Engineers</b>, <b>Teachers</b> and <b>Accountants</b>, with an average <b>Stress Level</b> between <b>3.9</b> and <b>4.6</b>.

### <b>Correlations</b>
Regarding the correlation between <b>Sleep Duration</b> and <b>Age</b>, we can observe that professionals under the age of 34 sleep less hours than the ones around 38 and older. 

<b>The Sleep Quality</b> also seems to improve in older individuals, with professionals under the average age of 34 being a minority in the chart.

In the correlation between <b>Sleep Duration</b> and <b>Sleep Quality</b>, it’s noticeable that the longer the individuals sleep, the better their sleep quality is.

<b>Daily Steps</b> and <b>Sleep Quality</b> show an obvious correlation. The less steps taken daily, the lower the quality of sleep is.

In terms of <b>Stress Levels</b>, we can see that the <b>Sleep Quality</b> goes down with higher Stress Levels. In one extreme, <b>Engineers</b> with a <b>Stress Level</b> under 4 show the best <b>Sleep Quality</b>, and <b>Sales Representatives</b> with a <b>Stress Level</b> of 8 report the worst <b>Sleep Quality</b>.

<b>Sleep Duration</b> and <b>Physical Activity</b> also show a noticeable correlation, where the individuals with a <b>Physical Activity</b> under 50 present a lower <b>Sleep Duration</b> than the rest.

At the same time, professionals that showed a higher <b>Physical Activity Level</b> present a better <b>Sleep Quality</b>. This makes sense, considering our third scatterplot graph shows a direct correlation between <b>Sleep Duration</b> and <b>Sleep Quality</b>.

### <b>BMI and Sleep Disorders</b>

By grouping the <b>Sleep Disorders</b> by <b>BMI Categories</b>, we can have a better look at the numbers and how these are related.

At first glance, we can see that the <b>Overweight</b> category presents the most amount of <b>Insomnia</b> and <b>Sleep Apnea</b>. What’s interesting is that the <b>Obese</b> category has less <b>Sleep Disorders</b> than the <b>Normal</b> one.

On the other side, <b>Normal</b> and <b>Normal Weight</b> have a better <b>Sleep Quality</b> and <b>Sleep Duration</b> than the <b>Obese</b> and <b>Overweight</b> categories, but not really by much difference.
